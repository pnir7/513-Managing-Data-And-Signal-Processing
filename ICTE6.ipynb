# Part1

```python
# Load the Iris dataset into a pandas dataframe

import pandas as pd

df = pd.read_csv('iris.csv')
print("Shape of the data: ", df.shape)
print("First 3 rows: ", df.head(3))
```

```python
# Print the number of observations, missing values, and NaN values
print("Number of observations: ", df.shape[0])
print("Missing values: ", df.isnull().sum().sum())
print("NaN values: ", df.isna().sum().sum())

```

```python
# Convert a 2-D array with ones on the diagonal and zeros elsewhere to a SciPy sparse matrix in CSR format

import numpy as np
from scipy.sparse import csr_matrix

arr = np.eye(3)
sparse_matrix = csr_matrix(arr)
print("Sparse Matrix in CSR format:\n", sparse_matrix)
```

```python
# View basic statistical details of the iris data
print("Statistical Details:\n", df.describe())
```

```python
# Print the observations of each species (setosa, versicolor, virginica) from the iris data
print("Observations of setosa: \n", df[df['Species'] == 'setosa'])
print("Observations of versicolor: \n", df[df['Species'] == 'versicolor'])
print("Observations of virginica: \n", df[df['Species'] == 'virginica'])
```

```python
# Drop the Id column from the dataframe and print the modified part
df = df.drop('Id', axis=1)
print("Dataframe after dropping Id column:\n", df.head(3))

```

```python
# Access the first four cells from the dataframe using the index and column labels

print("Accessing first four cells using index and column labels:")
print(df.loc[0, 'SepalLengthCm'])
print(df.loc[1, 'SepalWidthCm'])
print(df.loc[2, 'PetalLengthCm'])
print(df.loc[3, 'PetalWidthCm'])
```

# Part2

```python
# Create a plot to get a general statistics of the Iris data

import matplotlib.pyplot as plt

df.plot(kind='box', subplots=True, layout=(2,2), sharex=False, sharey=False)
plt.show()

```

```python
# Create a bar plot to get the frequency of the three species of the Iris data

species_count = df['Species'].value_counts()
species_count.plot(kind='bar')
plt.show()

```

```python
# Create a pie plot to get the frequency of the three species of the Iris data

species_count.plot(kind='pie', autopct='%1.1f%%')
plt.show()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset("iris")

# Create a scatter plot to see the relationship between petal length and width
sns.scatterplot(x="petal_length", y="petal_width", data=iris)
plt.show()

# Create a pairplot to see the distribution of Sepal length, Sepal width, Petal length, and Petal width
sns.pairplot(iris)
plt.show()
```

```python


# Create a joinplot using "hexbin" to describe individual distributions on the same plot between Sepal length and Sepal width
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind="hex")
plt.show()

# Create a joinplot using "kde" to describe individual distributions on the same plot between Sepal length and Sepal width
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind="kde")
plt.show()

# Create a joinplot and add regression and kernel density fits using "reg" to describe individual distributions on the same plot between Sepal length and Sepal width
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind="reg")
plt.show()
```

```python
# Draw a scatterplot, then add a joint density estimate to describe individual distributions on the same plot between Sepal length and Sepal width

import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid')

df = sns.load_dataset('iris')

sns.jointplot(x='sepal_length', y='sepal_width', data=df, kind='scatter')
plt.show()
```

```python
# Create a joinplot using "kde" to describe individual distributions on the same plot between Sepal length and Sepal width and use '+' sign as marker

sns.jointplot(x='sepal_length', y='sepal_width', data=df, kind='kde', marker='+')
plt.show()
```

```python
# Create a pairplot of the Iris data set and check which flower species seems to be the most separable

sns.pairplot(df, hue='species', palette='Dark2')
plt.show()
```

```python
# Find the correlation between variables of Iris data and create a heatmap using Seaborn to present their relations

corr = df.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python
# Create a box plot using Seaborn to present the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable of the Iris dataset

sns.boxplot(x='species', y='sepal_length', data=df)
plt.show()
```

```python
# Create a Principal Component Analysis (PCA) of the Iris dataset

from sklearn.decomposition import PCA
import numpy as np

X = df.drop('species', axis=1).values
y = df['species'].values

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

df_pca = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
df_pca['Species'] = y

sns.scatterplot(x='PC1', y='PC2', hue='Species', data=df_pca)
plt.show()
```
